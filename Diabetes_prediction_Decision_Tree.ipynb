```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
diabetes
```

```python
print(diabetes['DESCR'])
```

```python
x= pd.DataFrame(diabetes.data,columns = ['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
y=diabetes.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

### Hyperparameter tuning

```python
param = {'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
         'splitter': ['best', 'random'],
         'max_depth' : [1,2,3,4,10,15,20,25],
         'max_features': ['auto', 'sqrt', 'log2']
         }
```

```python
from sklearn.model_selection import GridSearchCV
regressor = DecisionTreeRegressor()
grid = GridSearchCV(regressor,param_grid=param,scoring='neg_mean_squared_error',cv=5)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python

grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
selectedmodel.fit(X_train,y_train)
```
